```python
#giving the student variable and reading the excel data
student_dfs=pd.read_csv("Students.csv",delimiter=';')
```

```python
#inspecting the variable data
student_dfs.head()
```

```python
#probability of the students to graduate
graduate_probability=student_dfs['Target'].value_counts(normalize=True).get('Graduate',0)
graduate_probability
```

```python
#probability of students to dropout of with material
dropout_with_marital_status_1=student_dfs[(student_dfs['Target']== 'Dropout') & (student_dfs['Marital status']==1)]
dropouts_proability=len(dropout_with_marital_status_1)/len(student_dfs)
dropouts_proability
```

```python
#The relation between mother and targetno C
relationsip=pd.crosstab(student_dfs['Mother\'s qualification'],student_dfs['Target'],normalize='columns')
relationsip
```
